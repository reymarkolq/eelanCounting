```python
from tensorflow.keras import Model
from tensorflow.keras.applications import MobileNetV3Small
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam

trainPath = "C:/Users/reymarkoliquino/OneDrive/Desktop/eeldiseases_dataset_train_for_model/train"
validPath = "C:/Users/reymarkoliquino/OneDrive/Desktop/eeldiseases_dataset_train_for_model/valid"

trainGenerator = ImageDataGenerator(
    rotation_range = 15, width_shift_range = 0.1,
    height_shift_range = 0.1, brightness_range = (0, 0.2)).flow_from_directory(trainPath, target_size=(256, 256), batch_size = 32)

validGenerator = ImageDataGenerator(
    rotation_range = 15, width_shift_range = 0.1,
    height_shift_range = 0.1, brightness_range = (0, 0.2)).flow_from_directory(validPath, target_size=(256, 256), batch_size = 32)

# Model building

baseModel = MobileNetV3Small(weights = "imagenet", include_top = False)

x = baseModel.output
x = GlobalAveragePooling2D()(x)
x = Dense(512, activation = 'relu')(x)
x = Dense(256, activation = 'relu')(x)
x = Dense(128, activation = 'relu')(x)

predictionLayer = Dense(3, activation = 'softmax')(x)

model = Model(inputs = baseModel.input, outputs = predictionLayer)

print(model.summary())
```

```python
# Freezing the layer of MobileNetV3 (From transfer learning)

for layer in model.layers[:-5]:
    layer.trainable = False
```

```python
# Compile

optimizer = Adam(learning_rate = 0.0001)
model.compile(optimizer="adam", loss = "categorical_crossentropy", metrics = ['accuracy'])
```

```python
# Train 

model.fit(trainGenerator, validation_data=validGenerator, epochs = 5)
```

```python
modelSavePath = "C:/Users/reymarkoliquino/OneDrive/Desktop/eeldiseases_dataset_train_for_model/EelV3.h5"
model.save(modelSavePath)
```
